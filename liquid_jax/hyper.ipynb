```python
from pathlib import Path
import pandas as pd
import optuna

def extract_results(folder: Path) -> pd.DataFrame:
    storage_path: str = f"sqlite:///{folder / 'study_le.db'}"
    
    study: optuna.Study = optuna.load_study(
        study_name="le",
        storage=storage_path
    )
    
    return study.trials_dataframe()

extract_results(Path(".") / "runs" / "hyper_le_20260413_190328")
```
